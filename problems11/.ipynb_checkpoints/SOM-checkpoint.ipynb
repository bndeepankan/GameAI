```python
def findDist(x, y):
    return np.sum((x - y) ** 2)
```

```python
def initGridSOM(matX, kx, ky):
    G = nx.generators.lattice.grid_2d_graph(kx, ky)
    G = nx.convert_node_labels_to_integers(G)
    
    m, n = matX.shape
    smpl = np.random.choice(n, kx * ky, replace = False)
    for i, v in enumerate(G):
        G.nodes[v]['w'] = matX[:, smpl[i]]
    
    return G
```

```python
def initSOM(matX, k, l=0):
    G = nx.generators.lattice.grid_2d_graph(k, 1, periodic = True)
    G = nx.convert_node_labels_to_integers(G, first_label = l)
    
    m, n = matX.shape
    smpl = np.random.choice(n, k, replace = False)
    for i, v in enumerate(G):
        G.nodes[v]['w'] = matX[:, smpl[i]]
    
    return G
```

```python
def trainSOM(matX, G, tmax = 25000, sigma0 = 1., eta0 = 1.):
    m, n = matX.shape
    
    matD = np.asarray(nx.floyd_warshall_numpy(G)) ** 2
    smpl = np.random.randint(0, n, size = tmax)
    for t in range(tmax):
        vecX = matX[:, smpl[t]]
        b = np.argmin([findDist(G.nodes[v]['w'], vecX) for v in G])
        eta = eta0 * (1. - t/tmax)
        sigma = sigma0 * np.exp(-t/tmax)
        for i,v in enumerate(G):
            h = np.exp(-0.5 * matD[b, i] / sigma**2)
            G.nodes[v]['w'] += eta * h * (vecX - G.nodes[v]['w'])
    
    return G
```

```python
def trainSOMV2(matX, G, tmax=1000, sigma0=1., eta0=1.): 
    m, n = matX.shape
    matW = np.vstack([G.nodes[v]['w'] for v in G.nodes()]).T 
    m, k = matW.shape
    matD = np.asarray(nx.floyd_warshall_numpy(G))**2 
    smpl = np.random.randint(0, n, size=tmax)
    for t in range(tmax):
    # NOTE: for all of the below to work, we must reshape the sampled column of X 
        vecX = matX[:,smpl[t]].reshape(m,1)
        b = np.argmin(np.sum((matW - vecX)**2, axis=0)) 
        eta = eta0 * (1. - t/tmax)
        sigma = sigma0 * np.exp(-t/tmax)
        vecH = np.exp(-0.5 * matD[b,:]/sigma**2) 
        matW += eta * vecH * (vecX - matW)
        for i, v in enumerate(G):
            G.nodes[v]['w'] = np.ravel(matW[:,i])
    return G
```

```python
def batchSOM(matX, G, tmax=10, sigma0=1.):
    m, n = matX.shape
    matD = np.asarray(nx.floyd_warshall_numpy(G)) ** 2
    for t in range(tmax):
        b = []
        vecX = []
        for j in range(n):
            vecX.append(matX[:, j])
            b.append(np.argmin([findDist(G.nodes[v]['w'], vecX[j]) for v in G]))
        sigma = sigma0 * np.exp(-t/tmax)
        for i,v in enumerate(G):
            h = []
            h_ = []
            for l in range(n):
                h.append(np.exp(-0.5 * matD[b[l], i] / sigma**2))
                h_.append(np.exp(-0.5 * matD[b[l], i] / sigma**2) * vecX[l])
            G.nodes[v]['w'] = sum(h_)/sum(h) 
    
    return G
```

```python
graph1 = initSOM(matY.T, 10, 0)
graph2 = initSOM(matY.T, 10, 10)
```
